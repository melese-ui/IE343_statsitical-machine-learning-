**(b) Fit the logistic regression model using a training data period from 1990 to 2008, with ````Lag2```` as the only predictor. Compute the confusion matrix and the overall fraction of correct predictions for the held out data (that is, the data from 2009 and 2010). Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.**

`````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score


train = df[df['Year'] < 2009]
test = df[df['Year'] >= 2009]


X_train = train[['Lag2']]  
y_train = train['Direction']

X_test = test[['Lag2']]
y_test = test['Direction']


logistic_sklearn = LogisticRegression(fit_intercept=True, C=1e9) 
logistic_sklearn_fit = logistic_sklearn.fit(X_train, y_train)


y_pred_test = logistic_sklearn_fit.predict(X_test)
conf_m = confusion_matrix(y_test, y_pred_test)


conf_m_df = pd.DataFrame(conf_m, columns=logistic_sklearn_fit.classes_, index=logistic_sklearn_fit.classes_)
conf_m_df.index.name = 'True'
conf_m_df.columns.name = 'Predicted'

# Displaying the confusion matrix
print(conf_m_df)



`````

From the confusion matrix, we have TN = 9, FN = 5,  TP = 56, FP = 34. When it's predicting Down, it predicted correctly 9 times and incorrectly 34 times out of 43 test sets that belong to 'Down'. So, the error rate with Down trends is 34/43 = 79.07%, which is high. For Up trends, it predicted correctly 56 times and incorrectly 5 times out of 61 test data sets that belong to 'Up'. So, the error rate with Up  trends is 8.2%, which is better compared to predicting Down trends. Overall, the logistic regression is at predicting Up trends rather than Down trends     

`````python
test_acc = accuracy_score(y_test, logistic_sklearn_fit.predict(X_test))
test_error = 1 - test_acc
print('Accuracy Using Logistic: {:.4f}'.format(test_acc))
print('Test Error Using Logistic: {:.4f}'.format(test_error))
`````

The overall fractions of correct predictions using Logistic regression is 0.625.

# 6. Newton's Method (20pt)

**In Lecture 3, we learned about Newton's method and two different perspectives of deriving the Newton's method. For this question, implement Newton's method.**

**[Tips] If you're unable to invert the Hessian Matrix due to it being singular, consider adding very small constants to the diagonal elements of the Hessian matrix.**

`````python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
from scipy.stats import logistic # Logistic function
import matplotlib.pyplot as plt



# load data
caravan = pd.read_csv('data/Caravan.csv', index_col=0)
caravan.Purchase = caravan.Purchase.astype('category')

X = caravan.iloc[:, 0:85].values

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

X = sm.add_constant(X)
y = caravan.Purchase.factorize()[0]

X_train = X[1001:]
y_train = y[1001:]
X_test = X[:1000]
y_test = y[:1000]

beta = np.random.rand(X.shape[1])
epochs = 20  # The number of iterations to perform gradient descent
n = float(len(X_train)) # Number of elements in X

cost = []

# Performing Newton's method
######################## Write your answere here ########################

# Regularize beta with a small value
beta *= 1e-15

# Perform optimization using Newton's method
for epoch in range(epochs):
    # Compute predicted probabilities using the logistic function
    predictions = logistic.cdf(np.dot(X_train, beta))
    
    # Compute the cross-entropy loss
    error = (-1 / n) * np.sum(y_train * np.log(predictions + 1e-15) + (1 - y_train) * np.log(1 - predictions + 1e-15))
    
    # Compute the gradient of the loss function
    gradient = (1 / n) * np.dot(X_train.T, predictions - y_train)
    
    # Compute the second derivative (Hessian) of the loss function
    sig_deriv = predictions * (1 - predictions)
    diagon_sig_deriv = np.diag(sig_deriv)
    second_derivative = np.linalg.inv((1 / n) * (X_train.T @ (diagon_sig_deriv @ X_train) + 1e-15))
    
    # Update beta using Newton's method
    beta -= np.dot(second_derivative, gradient)
    
    cost.append(error)



#########################################################################

plt.plot(np.arange(epochs), cost)
plt.title("Learning curve")
plt.xlabel("Epoch")
plt.ylabel(r"J($\beta$)");

plt.show()
y_test_pred = logistic.cdf(np.dot(X_test, beta)) > 0.5 # 0.5 should be fixed.
acc = accuracy_score(y_test, y_test_pred)
print("Accuracy of Logistic Regression (Newton's method): {}".format(acc))
`````
