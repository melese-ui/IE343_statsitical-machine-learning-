# 10. Mini-batch gradient descent (20pt)

**In the Lab 2, we learned how to implement gradient descent and stochastic gradient descent. For this question, implement mini-batch gradient descent. Set the ````batch_size```` to 10.**

`````python
from numpy.linalg import inv

boston = pd.read_csv('data/Boston.csv') # load data

X_train = boston[["lstat"]].values # regression using all input columns
X_train = sm.add_constant(X_train)
y_train = boston['medv'].values

beta_direct = inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
print(beta_direct)

beta = np.random.rand(2)
alpha = 0.001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent
n = float(len(X_train)) # Number of elements in X
batch_size = 10

######################## Write your answere here ########################
cost = []  # To store the cost at each epoch

for epoch in range(epochs):
    # Shuffle the data
    idx = np.random.permutation(len(X_train))
    X_train_shuffled = X_train[idx]
    y_train_shuffled = y_train[idx]

    # Mini-batch gradient descent
    for i in range(0, len(X_train), batch_size):
        X_batch = X_train_shuffled[i:i+batch_size]
        y_batch = y_train_shuffled[i:i+batch_size]
        
        # Compute the gradient
        gradient = -(2/n) * X_batch.T.dot(y_batch - X_batch.dot(beta))
        
        # Update the parameters
        beta -= alpha * gradient
        
    # Compute the cost
    J = np.mean((X_train.dot(beta) - y_train)**2)
    cost.append(J)


#########################################################################

print("Learned parameters through SGD: {}".format(beta))
print("Directly learned parameters: {}".format(beta_direct))   
plt.plot(np.arange(epochs), cost)
plt.title("Learning curve")
plt.xlabel("Epoch")
plt.ylabel(r"J($\beta$)");
`````
